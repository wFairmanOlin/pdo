### Numpy

Numpy is a core library that provides an *array* data structure. These arrays are useful for representing *vectors* in our computations.

Some *quick facts* about accessing numpy arrays:

- Python is a *zero-index* language, meaning you use `0` to access the first element, `1` for the second, etc.
- Python provides the special syntax `[-1]` to access the *last* entry in an array.
- Numpy arrays can be *multi-dimensional*; a one-dimensional array is a vector, while a two-dimensional array is a matrix.
  - For a Numpy matrix `A`, we access entires with a *multi-index* `A[i, j]`
  - The first entry `i` accesses the row number
  - The second entry `j` accesses the column number
  - The special syntax `[:]` returns all the entries along an axis; for instance `A[0, :]` returns the first *row* of the matrix `A`
- Numpy arrays enable numerous vector operations:
  - Addition `v1 + v2`
  - Subtraction `v1 - v2`
  - Inner product `np.dot(v1, v2)`
  - Length `np.linalg.norm(v)`
  
*Note*: We'll use the terms *array* and *vector* somewhat interchangeably in this course. Note that technically we use arrays to *represent* vectors.

$$c\text{-contour} \equiv \{x, y\,|\,f(x, y) = c\}.$$

### Case Study: Minimum-energy spring configuration

We'll use the following problem as a warm-up for some fundamentals:

> Let $v_1, v_2, v_3$ be *anchor points* in 2d space, each of which is connected to a single test point $x$ by a spring with spring constant $K_i$ and zero equilibrium length. What is the point $x^*$ that minimizes the energy of the system?

The principle of least-energy implies that this minimum-energy point $x^*$ will be the location where the test point will settle. Our first step will be to translate this problem statement into an optimization problem.

The following code visualizes the problem schematically for a specific choice of anchor points.


````python
# NO NEED TO EDIT; schematic diagram
# Anchor points
V = np.array([
    [1, 3],
    [2, 1],
    [4, 4],
])
# Test point
x = [1.6, 2.0]

# Visualize
for i in range(V.shape[0]):
    plt.plot([V[i, 0], x[0]], [V[i, 1], x[1]], 'k-')
    
plt.plot(V[:, 0], V[:, 1], "b.", label="anchor")
plt.plot(x[0], x[1], "r.", label="test point")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc=0)
````

*Aside*: This is a simple problem, but similar structural layouts are used in practice. For instance the [Arecibo Observatory](https://en.wikipedia.org/wiki/Arecibo_Observatory) was suspended with a cable system not too unlike this layout!


The first step in solving a problem with optimization is to *translate* it into an optimization problem! the *standard form* for an optimization problem is the following:

$$\min f(x)$$
$$\text{wrt} \,x$$
$$\text{s.t.} \,g(x) \leq 0$$

Each of these pieces is important:

- $\min f(x)$ is the *objective*; in this case minimize the function $f(x)$.
- $\text{wrt} \,x$ stands for *with respect to* $x$; it is important to *specify* which variables are *active* during our optimization.
  - *Note*: The "wrt" line defines, by omission, which inputs are *constants* during optimization.
- $\text{s.t.} \,g(x) \leq 0$ denotes a *constraint*; "s.t." is short for *subject to*. Not all optimization problems have constraints.

*Aside*: An optimization problem with constraints *but no objective* is called a *feasibility problem*.


### __q2.1__ Translate the problem

Translate the minimum-energy spring problem into standard form. **Don't forget to specify the "wrt" variables!**

**Task**:

(Write your translated optimization problem here; you don't have to use LaTeX if you don't want to.)

$$\min \Sigma{E_i(x^*)}$$
$$\text{wrt} \,x^*$$

Once you've derived your objective function $c(x)$, implement it below to check your work.

*Hint 1*: Recall that the energy energy of a spring is given by $E_i = \frac{1}{2} K_i \Delta e_i^2$ where $K_i$ is the spring constant and $\Delta e_i$ is the extension of the spring.  

*Hint 2*: Since the springs have zero equilibrium length, their extention is given by the distance between points, that is $\Delta e_i = \|x - v_i\|_2$, where $\|x - v\|_2 = \sqrt{(x_1 - v_1)^2 + (x_2 - v_2)^2}$ is the ordinary (Euclidean) distance.

*Hint 3*: The numpy function `np.linalg.norm(d)` implements the math operator $\|d\|_2$. You're better off using this **vector operator** rather than using the individual entires of $x$.

*Hint 4*: The exponential operator in Python is **not** `^`. Rather it is a double-star `x**2`.

````python
###
# TASK: Implement your objective function below
# NOTE: Your function should have the signature objective(X)
###

# NO NEED TO EDIT; use these constants in your function
# Spring constants
K = [1, 2, 3]
# Anchor points
V = np.array([
    [1, 3],
    [2, 1],
    [4, 4],
])

# TASK: Implement the objective (energy expression)
def objective(X):
    # Hint: This one is easier if you *do not* 
    # unpack X (use np.linalg.norm)
    e_sum = 0
    for i in range(len(K)):
        e_sum += .5 * K[i] * np.linalg.norm(X-V[i])**2
    
    return e_sum



print(objective([0,0]))
# NO NEED TO EDIT; this checks your answer
res_obj = objective([0, 0])
assert(abs(res_obj - 58) < 1e-6)
print("Success!")
````

It is a *necessary* condition for the minimum value of a function to occur at a point where the *gradient* of the function equals zero. We'll use the approach of setting the gradient of our objective to zero to solve this problem.

### __q2.2__ Derive the gradient

Derive an expression for the gradient of the energy $\nabla E(x)$.

Once you've derived the gradient, implement it as a function below to check your work.

*Hint*: Note that $\|x - v\|_2^2 = (x - v)^{\top} (x - v)$ (the inner product of two vectors), and $\nabla_x \left((x - v)^{\top} (x - v)\right) = 2 (x - v)$ (product rule applied to vectors). You can *greatly* simply your work if you use these vector expressions.

````python
###
# TASK: Implement your gradient function below
# NOTE: Your function should have the signature gradient(X)
###

# NO NEED TO EDIT; use these constants in your function
# Spring constants
K = [1, 2, 3]
# Anchor points
V = np.array([
    [1, 3],
    [2, 1],
    [4, 4],
])

# TODO: Implement your gradient function
def gradient(X):
    # HINT: This one is *much easier* if you do not unpack X!
    return np.dot(K,(X-V))



# NO NEED TO EDIT; this checks your answer
res_grad = gradient([0, 0])
assert(abs(res_grad[0] + 17) < 1e-6)
assert(abs(res_grad[1] + 17) < 1e-6)

print("Success!")

````

As noted above it is a *necessary* condition for the gradient of a function to be zero in order for a point to be the minimum value (in the unconstrained case). A point with $\nabla f(x^*) = 0$ is called a *stationary point* $x^*$, and under certain conditions a stationary point is also an optimal point. It happens that the minimum-energy spring problem satisfies such conditions, so we can find the minimum energy point by finding the stationary point.

### __q2.3__ Find the stationary point

Solve for $x^*$ via $\nabla E(x^*) = 0$.



Once you've derived the stationary point $x^*$ in terms of an analytic expression, implement your solution below to check your work.


````python
###
# TASK: Set the gradient equal to zero, use this to solve for x_star
###

# NO NEED TO EDIT; use these constants in your function
# Spring constants
K = [1, 2, 3]
# Anchor points
V = np.array([
    [1, 3],
    [2, 1],
    [4, 4],
])

# TASK: Implement your solution as a function solve(K)
def solve(K):
    x = [0,0]
    k_sum = 0
    kv_sum = 0
    for i in range(len(K)):
        k_sum += K[i]
        kv_sum += np.dot(K[i],V[i])
    return kv_sum/k_sum

x_star = solve(K)

# NO NEED TO EDIT; use the following code to check & visualize your work
print("q2.3 Diagnostics")
print("x_star =", x_star)
print("gradient =", gradient(x_star))
print()

assert(abs(x_star[0] - 2.833333) < 1e-6)
assert(abs(x_star[1] - 2.833333) < 1e-6)

# Visualize
for i in range(V.shape[0]):
    plt.plot([V[i, 0], x_star[0]], [V[i, 1], x_star[1]], 'k-')
    
plt.plot(V[:, 0], V[:, 1], "b.", label="anchor")
plt.plot(x_star[0], x_star[1], "r.", label="test point")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc=0)
````

### __q2.4__ Use the solution to do useful work

In the following I assume the spring constants $K$ can vary independently according to $K_i = 0.5 \pm 0.25$. I model the uncertainty in the spring constants probabilistically, and perform a Monte Carlo analysis to assess where the minimum-energy point tends to land based on this model. Your task is to interpret the results (see below).

````python
###
# TASK: Run this code and interpret the results
###

# NO NEED TO EDIT THIS; run and interpret the results
# Construct Grama model using your function
md_point = (
    gr.Model("Minimum-energy point")
    >> gr.cp_function(
        fun=solve,
        var=["K1", "K2", "K3"],
        out=["x", "y"],
    )
    >> gr.cp_marginals(
        K1=dict(dist="uniform", loc=0.25, scale=0.5),
        K2=dict(dist="uniform", loc=0.25, scale=0.5),
        K3=dict(dist="uniform", loc=0.25, scale=0.5),
    )
    >> gr.cp_copula_independence()
)

# Generate data from model
df_results = (
    md_point
    >> gr.ev_monte_carlo(n=2e3, df_det="det", seed=101)
)


# Visualize data
fig, ax = plt.subplots()
hb = ax.hexbin(df_results.x, df_results.y, gridsize=15, cmap="Purples")
cb = fig.colorbar(hb, ax=ax)
ax.plot(V[:, 0], V[:, 1], "r.", label="anchor")

cb.set_label('counts')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Distribution of minimum-energy points")
plt.legend(loc=0)
````

````python
# calculating probability of x being within x* +- .1
K = [.5, .5, .5]
x_star = solve(K)
x_star = [2.33, 2.67]
print("x_star", x_star)
circle = plt.Circle(x_star, np.sqrt(.1**2 + .1**2), color='r', fill=False)
fig, ax = plt.subplots()
ax.scatter(df_results.x, df_results.y, s=1)
ax.add_patch(circle)

count = 0
for i in range(2000):
#     print(df_results.x[i])
    if abs(df_results.x[i] - x_star[0]) < .1:
        if abs(df_results.y[i] - x_star[1]) < .1:
            count += 1

print("fell within .1 of x and y coordinates: ", count)
print("prob success: ", count/2000)
````
